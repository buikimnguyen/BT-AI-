```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
BloodSugarLevel = ctrl.Antecedent(np.arange(0, 300.1, 0.1), 'BloodSugarLevel')
BMI = ctrl.Antecedent(np.arange(0, 40.1, 0.1), 'BMI')
FrequentUrination = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'FrequenUrination')
FatigueLevel = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'FatigueLevel')
RiskLevel = ctrl.Consequent(np.arange(0, 10.1, 0.1), 'RiskLevel')
```

```python
BloodSugarLevel['low'] = fuzz.trimf(BloodSugarLevel.universe, [0, 0, 80])
BloodSugarLevel['normal'] = fuzz.trimf(BloodSugarLevel.universe, [80, 100, 120])
BloodSugarLevel['high'] = fuzz.trimf(BloodSugarLevel.universe, [120, 160, 200])
BloodSugarLevel['very_high'] = fuzz.trimf(BloodSugarLevel.universe, [200, 200, 300])
BMI['skinny'] = fuzz.trimf(BMI.universe, [0, 0, 18.5])
BMI['normal'] = fuzz.trimf(BMI.universe, [18.5, 21.5, 24.9])
BMI['overweight'] = fuzz.trimf(BMI.universe, [24.9, 27.5, 29.9])
BMI['obesity'] = fuzz.trimf(BMI.universe, [30, 30, 40])
FrequentUrination['less'] = fuzz.trimf(FrequentUrination.universe, [0, 0, 5])
FrequentUrination['medium'] = fuzz.trimf(FrequentUrination.universe, [4, 6 ,8])
FrequentUrination['much'] = fuzz.trimf(FrequentUrination.universe, [7, 10, 10])
FatigueLevel['low'] = fuzz.trimf(FatigueLevel.universe, [0, 0, 4])
FatigueLevel['medium'] = fuzz.trimf(FatigueLevel.universe, [3, 5.25, 7.5])
FatigueLevel['high'] = fuzz.trimf(FatigueLevel.universe, [7, 10, 10])
RiskLevel['low'] = fuzz.trimf(RiskLevel.universe, [0, 0, 3])
RiskLevel['medium'] = fuzz.trimf(RiskLevel.universe, [2, 4, 6])
RiskLevel['high'] = fuzz.trimf(RiskLevel.universe, [4, 6, 8])
RiskLevel['very_high'] = fuzz.trimf(RiskLevel.universe, [7, 8, 10])
```

```python
BloodSugarLevel.view()
```

```python
FatigueLevel.view()
```

```python
RiskLevel.view()
```

```python
Rule1 = ctrl.Rule(BloodSugarLevel['very_high'] & FrequentUrination['much'], RiskLevel['very_high'])
Rule2 = ctrl.Rule(BloodSugarLevel['very_high'] & FrequentUrination['medium'], RiskLevel['high'])
Rule3 = ctrl.Rule(BloodSugarLevel['very_high'] & FrequentUrination['less'], RiskLevel['medium'])
Rule4 = ctrl.Rule(BloodSugarLevel['high'] & FrequentUrination['much'], RiskLevel['very_high'])
Rule5 = ctrl.Rule(BloodSugarLevel['high'] & FrequentUrination['medium'], RiskLevel['high'])
Rule6 = ctrl.Rule(BloodSugarLevel['high'] & FrequentUrination['less'], RiskLevel['medium'])
Rule7 = ctrl.Rule(BloodSugarLevel['normal'] & FrequentUrination['much'], RiskLevel['high'])
Rule8 = ctrl.Rule(BloodSugarLevel['normal'] & FrequentUrination['medium'], RiskLevel['medium'])
Rule9 = ctrl.Rule(BloodSugarLevel['normal'] & FrequentUrination['less'], RiskLevel['low'])
Rule10 = ctrl.Rule(BloodSugarLevel['low'] & FrequentUrination['much'], RiskLevel['high'])
Rule11 = ctrl.Rule(BloodSugarLevel['low'] & FrequentUrination['medium'], RiskLevel['medium'])
Rule12 = ctrl.Rule(BloodSugarLevel['low'] & FrequentUrination['less'], RiskLevel['low'])

Rule13 = ctrl.Rule(BloodSugarLevel['very_high'] & BMI['obesity'], RiskLevel['very_high'])
Rule14 = ctrl.Rule(BloodSugarLevel['very_high'] & BMI['overweight'], RiskLevel['very_high'])
Rule15 = ctrl.Rule(BloodSugarLevel['very_high'] & BMI['normal'], RiskLevel['high'])
Rule16 = ctrl.Rule(BloodSugarLevel['very_high'] & BMI['skinny'], RiskLevel['high'])
Rule17 = ctrl.Rule(BloodSugarLevel['high'] & BMI['obesity'], RiskLevel['very_high'])
Rule18 = ctrl.Rule(BloodSugarLevel['high'] & BMI['overweight'], RiskLevel['high'])
Rule19 = ctrl.Rule(BloodSugarLevel['high'] & BMI['normal'], RiskLevel['medium'])
Rule20 = ctrl.Rule(BloodSugarLevel['high'] & BMI['skinny'], RiskLevel['medium'])
Rule21 = ctrl.Rule(BloodSugarLevel['normal'] & BMI['obesity'], RiskLevel['high'])
Rule22 = ctrl.Rule(BloodSugarLevel['normal'] & BMI['overweight'], RiskLevel['medium'])
Rule23 = ctrl.Rule(BloodSugarLevel['normal'] & BMI['normal'], RiskLevel['low'])
Rule24 = ctrl.Rule(BloodSugarLevel['normal'] & BMI['skinny'], RiskLevel['low'])
Rule25 = ctrl.Rule(BloodSugarLevel['low'] & BMI['obesity'], RiskLevel['medium'])
Rule26 = ctrl.Rule(BloodSugarLevel['low'] & BMI['overweight'], RiskLevel['medium'])
Rule27 = ctrl.Rule(BloodSugarLevel['low'] & BMI['normal'], RiskLevel['low'])
Rule28 = ctrl.Rule(BloodSugarLevel['low'] & BMI['skinny'], RiskLevel['low'])

Rule29 = ctrl.Rule(BloodSugarLevel['very_high'] & FatigueLevel['high'], RiskLevel['very_high'])
Rule30 = ctrl.Rule(BloodSugarLevel['very_high'] & FatigueLevel['medium'], RiskLevel['high'])
Rule31 = ctrl.Rule(BloodSugarLevel['very_high'] & FatigueLevel['low'], RiskLevel['medium'])
Rule32 = ctrl.Rule(BloodSugarLevel['high'] & FatigueLevel['high'], RiskLevel['very_high'])
Rule33 = ctrl.Rule(BloodSugarLevel['high'] & FatigueLevel['medium'], RiskLevel['high'])
Rule34 = ctrl.Rule(BloodSugarLevel['high'] & FatigueLevel['low'], RiskLevel['medium'])
Rule35 = ctrl.Rule(BloodSugarLevel['normal'] & FatigueLevel['high'], RiskLevel['high'])
Rule36 = ctrl.Rule(BloodSugarLevel['normal'] & FatigueLevel['medium'], RiskLevel['medium'])
Rule37 = ctrl.Rule(BloodSugarLevel['normal'] & FatigueLevel['low'], RiskLevel['low'])
Rule38 = ctrl.Rule(BloodSugarLevel['low'] & FatigueLevel['high'], RiskLevel['medium'])
Rule39 = ctrl.Rule(BloodSugarLevel['low'] & FatigueLevel['medium'], RiskLevel['low'])
Rule40 = ctrl.Rule(BloodSugarLevel['low'] & FatigueLevel['low'], RiskLevel['low'])
```

```python
risk_ctrl = ctrl.ControlSystem([Rule1, Rule2, Rule3,Rule4,Rule5,Rule6,Rule7,Rule8,Rule9,Rule10,Rule11,Rule12,Rule13,Rule14,Rule15,Rule16,Rule17,Rule18,Rule19,Rule20,Rule21,Rule22,Rule23,Rule24,Rule25,Rule26,Rule27,Rule28,Rule29,Rule30,Rule31,Rule32,Rule33,Rule34,Rule35,Rule36,Rule37,Rule38,Rule39,Rule40])
risk = ctrl.ControlSystemSimulation(risk_ctrl)
risk.input['BloodSugarLevel'] = 154
risk.input['BMI'] = 24
risk.input['FrequenUrination'] = 4
risk.input['FatigueLevel'] = 4
risk.compute()
print(risk.output['RiskLevel'])
RiskLevel.view(sim=risk)
```
